```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from copy import deepcopy
import time
```

```python
def padding(im,filt_sz,pad_type = 'constant'):
    f_h,f_w = filt_sz
    if (f_h-1)%2 != 0 :
        pad_r = ((f_h-1)//2+1,(f_h-1)//2)
    else :
        pad_r = ((f_h-1)//2,(f_h-1)//2)
    if (f_w-1)%2 != 0 :
        pad_w = ((f_w-1)//2+1,(f_w-1)//2)
    else :
        pad_w = ((f_w-1)//2,(f_w-1)//2)
    pad_sz = (pad_r,pad_w)
    im1 = np.pad(im,pad_sz,mode = pad_type)
    return im1
```

```python
def conv(im, kernel):
    im = padding(im, kernel.shape)
    filtered_im = np.zeros((im.shape[0] - kernel.shape[0] + 1, im.shape[1] - kernel.shape[1] + 1))
    for w in range(im.shape[1] - kernel.shape[1]):
        for h in range(im.shape[0] - kernel.shape[0]):
            filtered_im[h,w] = np.sum(im[h:h + kernel.shape[1], w:w + kernel.shape[0]]*kernel)
    return filtered_im

def col_for_conv(im, kernel):
    k_h,k_w = kernel.shape
    im_h,im_w = im.shape
    a = np.transpose(((np.arange(im_h - k_h+1)[:,None] + np.arange(im_w - k_w+1)*im_h).ravel(order='F')[:,None]+ np.arange(k_h))[:,:,None],(1,2,0))
    a = np.reshape(a+im_h*np.arange(k_w)[None,:,None],(k_h*k_w,-1),order='F')
    return im.ravel(order='F')[a]

def fasterconv(im, kernel, do_padding = True, padding_type = 'constant'):
    if do_padding:
        im = padding(im, kernel.shape, padding_type)
    k_h,k_w = kernel.shape
    im_h,im_w = im.shape
    col_im = col_for_conv(im, kernel)
    kernel_col = kernel.ravel()[:,None]
    return np.reshape(np.sum(col_im*kernel_col, axis=0), (im_h - k_h+1, im_w - k_w+1) ,order='F')

def avg_filter(im,k):
    kernel = np.ones((k,k))/k**2
    return fasterconv(im, kernel, False).astype(int)  

def faster_avg_filter(im, k):
    kernel = np.ones((1,k))/k
    return fasterconv(fasterconv(im,kernel, False),kernel.T, False)
```

```python
# reading images
noisy = cv2.imread("../images/Noisy.jpg", 0)

#applying filter
un_noisy = faster_avg_filter(noisy, 5)

#plotting image
f, ax = plt.subplots(1, 2, figsize=(10,10))
[axi.set_axis_off() for axi in ax[:].ravel()]
ax[0].title.set_text("given image") ;ax[1].title.set_text("after applying mean filter")
ax[0].imshow(noisy, 'gray')
ax[1].imshow(un_noisy, 'gray'); plt.show()
```

```python
# def median_filt(im,k):
#     im_h, im_w = im.shape
#     col_img = col_for_conv(im, np.zeros((k,k)))
#     return np.median(col_img,axis=0).reshape((im_h - k+1, im_w - k+1) ,order='F')

def median_filt(im,k):
    filtered_im = np.zeros((im.shape[0] - k + 1, im.shape[1] - k + 1))
    for w in range(im.shape[1] - k):
        for h in range(im.shape[0] - k):
            filtered_im[h,w] = np.median(im[h:h + k, w:w + k].reshape((1,k**2)))
    return filtered_im

def efficient_median_filt(im,k):
    im_h, im_w = im.shape
    col_img = col_for_conv(im, np.zeros((k,k)))
    ret = np.zeros((col_img.shape[1],1))
    for i in range(col_img.shape[1]):
        if i%(im_h - k + 1) == 0:
            running_hist = np.histogram(col_img[:,i],np.arange(257))[0]
            ret[i] = np.median(col_img[:,i], axis=0)
        else:
            running_hist[col_img[:,i-1][np.arange(0,k**2,k)]] = running_hist[col_img[:,i-1][np.arange(0,k**2,k)]] - 1
            running_hist[col_img[:,i][np.arange(k-1,k**2,k)]] = running_hist[col_img[:,i][np.arange(k-1,k**2,k)]] + 1
            csum =  np.cumsum(running_hist)
            ret[i] = np.argmin(csum < csum[-1]/2)
    return ret.reshape(((im_h-k+1),(im_w-k+1)), order = 'F')
```

```python
def prewitt(im, k):
    if k == 3:
        kernel_x = np.zeros((k,k))
        kernel_x[:,0] = -1; kernel_x[:,-1] = 1
        kernel_y = deepcopy(-1*kernel_x.T).astype(int)
    elif k==5:
        kernel_x = np.zeros((k,k))
        kernel_x[:,0] = -2; kernel_x[:,1] = -1;kernel_x[:,-1] = 2;kernel_x[:,-2] = 1;
        kernel_y = deepcopy(kernel_x.T).astype(int)
        print("prewitt 5 Gx filter should look like: \n", kernel_x)
        print("prewitt 5 Gy filter should look like: \n", kernel_y)
        
    edge_x = fasterconv(im, kernel_x)
    edge_y = fasterconv(im, kernel_y)
    return edge_x, edge_y

def sobel(im, k):
    if k == 3:
        kernel_x = np.zeros((k,k))
        kernel_x[:,0] = -1; kernel_x[k//2,0] = -2; kernel_x[:,-1] = -1*kernel_x[:,0]
        kernel_y = deepcopy(-1*kernel_x.T).astype(int)
    elif k==5:
        kernel_x = np.zeros((k,k))
        kernel_x[:,0] = -2; kernel_x[k//2,0] = -4; kernel_x[:,-1] = -1*kernel_x[:,0]
        kernel_x[:,1] = -1; kernel_x[k//2,1] = -2; kernel_x[:,-2] = -1*kernel_x[:,1]
        kernel_y = deepcopy(-1*kernel_x.T).astype(int)
        print("sobel 5 Gx filter should look like: \n", kernel_x)
        print("sobel 5 Gy filter should look like: \n", kernel_y)
        
    edge_x = fasterconv(im, kernel_x)
    edge_y = fasterconv(im, kernel_y)
    return edge_x, edge_y

def roberts(im):
    kernel_x = np.array([[0,1],[-1,0]]); kernel_y = np.array([[1,0],[0,-1]])
    edge_x = fasterconv(im, kernel_x)
    edge_y = fasterconv(im, kernel_y)
    return edge_x, edge_y

def laplacian(im, k_type = 1):
    if k_type == 1:
        kernel = np.array([[0,1,0],[1,-4,1],[0,1,0]])
    else:
        kernel = np.array([[1,1,1],[1,-8,1],[1,1,1]])
    return fasterconv(im, kernel)
```

```python
def bilateral_filt(im, k, sig_s, sig_r):
    im_h, im_w = im.shape
    ind = np.indices((k,k)) - k//2
    G_s = np.exp(-(ind[0]**2 + ind[1]**2)/(2*sig_s**2)).reshape(k*k,1, order = 'F'); G_s = G_s/np.sum(G_s)
    col_im = col_for_conv(im, np.zeros((k,k)))
    shifted_intensities = col_im - col_im[col_im.shape[0]//2,:]
    G_r = np.exp((-shifted_intensities**2)/(2*sig_r**2))
    filtered_im = np.sum(col_im*(G_r*G_s)/np.sum(G_r*G_s, axis=0), axis=0)
    return np.reshape(filtered_im, (im_h - k+1, im_w - k+1), order='F')

def bilateral_filt_colour(im, k, sig_s, sig_r):
    f1 = bilateral_filt(padding(im[:,:,0], (k,k)), k, sig_s, sig_r)
    f2 = bilateral_filt(padding(im[:,:,1], (k,k)), k, sig_s, sig_r)
    f3 = bilateral_filt(padding(im[:,:,2], (k,k)), k, sig_s, sig_r)
    return np.dstack((np.dstack((f1,f2)),f3))    
```

```python
def cross_bilateral_filt(im_r, im_s, k, sig_s, sig_r):
    im_h_r, im_w_r = im_r.shape
    im_h_s, im_w_s = im_s.shape
    ind = np.indices((k,k)) - k//2
    G_s = np.exp(-(ind[0]**2 + ind[1]**2)/(2*sig_s**2)).reshape(k*k,1, order = 'F'); G_s = G_s/np.sum(G_s)
    col_im_r = col_for_conv(im_r, np.zeros((k,k)))
    col_im_s = col_for_conv(im_s, np.zeros((k,k)))
    shifted_intensities = col_im_r - col_im_r[col_im_r.shape[0]//2,:]
    G_r = np.exp((-shifted_intensities**2)/(2*sig_r**2))
    filtered_im = np.sum(col_im_s*(G_r*G_s)/np.sum(G_r*G_s, axis=0), axis=0)
    return np.reshape(filtered_im, (im_h_s - k+1, im_w_s - k+1), order='F')

def cross_bilateral_filt_colour(im_r, im_s, k, sig_s, sig_r):
    f1 = cross_bilateral_filt(padding(im_r[:,:,0], (k,k)), padding(im_s[:,:,0], (k,k)), k, sig_s, sig_r)
    f2 = cross_bilateral_filt(padding(im_r[:,:,1], (k,k)), padding(im_s[:,:,1], (k,k)), k, sig_s, sig_r)
    f3 = cross_bilateral_filt(padding(im_r[:,:,2], (k,k)), padding(im_s[:,:,2], (k,k)), k, sig_s, sig_r)
    return np.dstack((np.dstack((f1,f2)),f3))  

def weighted_avg(im1, im2, alpha):
    im1 = cv2.resize(im1, im2.shape[-2::-1])
    return alpha*im1 + (1-alpha)*im2
```

```python
#reading images
cake_flash = cv2.cvtColor(cv2.imread("../images/cake_flash.jpg"), cv2.COLOR_BGR2RGB)
cake_no_flash = cv2.cvtColor(cv2.imread("../images/cake_noflash.jpg"), cv2.COLOR_BGR2RGB)

#applying filter
filtered_cake = cross_bilateral_filt_colour(cake_flash/255, cake_no_flash/255, 7, 6, 11)
good_cake = weighted_avg(cake_flash/255, filtered_cake, 0.3)

f, ax = plt.subplots(1, 2, figsize=(10,10))
[axi.set_axis_off() for axi in ax[:].ravel()]
ax[0].title.set_text("given image") ;ax[1].title.set_text("after applying cross bilateral filter")
ax[0].imshow(cake_flash, 'gray')
ax[1].imshow(good_cake, 'gray'); plt.show()
```

```python
def adaptive_thresholding(im, k):
    im = padding(im,(k,k))
    col_im = col_for_conv(im, np.zeros((k,k)))
    thresh = np.mean(col_im, axis = 0)
    return np.reshape(((col_im[k**2//2,:]>=thresh).astype(int)),(im.shape[0]-k+1, im.shape[1]-k+1), order='F')
    
def make_cartoon(im, sig):
    filtered = (255*bilateral_filt_colour(im/255, 5, 6, 11)).astype(int)
    ind = np.indices((5,5)) - 5//2
    gaussian_kernel = np.exp(-(ind[0]**2 + ind[1]**2)/(2*sig**2)); gaussian_kernel = gaussian_kernel/np.sum(gaussian_kernel)
    im_cpy = cv2.cvtColor(deepcopy(im),cv2.COLOR_BGR2GRAY)
    edges = (fasterconv(im_cpy, gaussian_kernel) - im_cpy)
    edges = adaptive_thresholding(im_cpy, 23)
#     threshold = (np.min(edges)+np.max(edges))/2
#     edges = edges>threshold; edges = np.dstack((np.dstack((edges, edges)), edges))
    edges = np.dstack((np.dstack((edges, edges)), edges))
    
    return ((255*edges).astype(int))&filtered
```

```python
def dft(im, sz):
    # could have vectorized this too but the space complexity would have gone up to (sz C 2)*imh*imw
    imh, imw = im.shape
    ind = np.indices(im.shape) 
    ind[0] = ind[0] - im.shape[0]//2; ind[1] = ind[1] - im.shape[1]//2
    dft = np.zeros((sz,sz),dtype=complex)
    for i in range(sz):
        for k in range(sz):
            dft[i,k] = np.sum(im*np.exp(-2*np.pi*1j*(i*ind[0]/imh + k*ind[1]/imw)))
    return dft
```

```python
def fft_recurr(x):
    n = np.uint(x.shape[0]) 
    if n < 2:
        return x
    
    even = fft_recurr(x[0:n:2,:])
    odd = fft_recurr(x[1:n:2,:])
    W = np.exp(-np.pi*2j*np.arange(n//2)/n)[:,None]
    return np.concatenate((even + W*odd, even - W*odd),axis=0)

def fft(x):
    n = x.shape[0]
    padsz = np.int(2**np.ceil(np.math.log(n,2)) - n)
    x1 = np.pad(x,((0,padsz),(0,0)),'constant')
    
    return fft_recurr(x1)

def fft2(x):
    return fft(fft(x).T).T

def ifft_recurr(x):
    n = np.uint(x.shape[0]) 
    if n < 2:
        return x
    
    even = ifft_recurr(x[0:n:2,:])
    odd = ifft_recurr(x[1:n:2,:])
    W = np.exp(np.pi*2j*np.arange(n//2)/n)[:,None]
    return np.concatenate((even + W*odd, even - W*odd),axis=0)

def ifft(x):
    n = x.shape[0]
    padsz = np.int(2**np.ceil(np.math.log(n,2)) - n)
    x1 = np.pad(x,((0,padsz),(0,0)),'constant')
    
    return ifft_recurr(x1)

def ifft2(x):
    return ifft(ifft(x).T).T
```

```python
def gaussian_blur(im, k, sig):
    ind = np.indices((k,k)) - k//2
    gaussian_kernel = np.exp(-(ind[0]**2 + ind[1]**2)/(2*sig**2)); gaussian_kernel = gaussian_kernel/np.sum(gaussian_kernel)
    return fasterconv(im, gaussian_kernel, padding_type = 'edge')

def highboost_filt(im, k, sig, alpha):
    gaussian_img = gaussian_blur(im, k, sig)
    edges = abs(gaussian_img - im)
    filtered_img = im + alpha*edges
    return filtered_img/np.max(filtered_img)
```

```python
noisy_lena = cv2.cvtColor(cv2.imread("../images/noisy_lena.png"), cv2.COLOR_BGR2GRAY)

lenna_fft = np.fft.fftshift(fft2(noisy_lena))
mask = np.ones((lenna_fft.shape))
mask[230:270, :237] = 0; mask[230:270, 275:] = 0
mask = gaussian_blur(mask, 9, 15)
lenna_fft = lenna_fft*mask

lena = abs(ifft2(lenna_fft))
lena = highboost_filt(lena, 3, 3, 0.3)

f, ax = plt.subplots(1, 2, figsize=(8,8))
[axi.set_axis_off() for axi in ax[:].ravel()]
ax[0].title.set_text("noisy lena") ;ax[1].title.set_text("denoised lena")
ax[0].imshow(noisy_lena, 'gray')
ax[1].imshow(lena, cmap = 'gray'); plt.show()
```

Since the noisy image has repeating patterns in x direction, therefore we can mask out the x axis frequences to get a better image. To avoid the ringing effect, we can smooth the mask applied to fourier domain. Then I have performed highboost filtering to enhance the edges lost by removing the high frequencies in the image.
