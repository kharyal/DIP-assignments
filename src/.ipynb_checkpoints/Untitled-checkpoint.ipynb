```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# reading images

ims = []

for i in range(4):
    ims.append(cv2.imread(str(i+1)+".png"))
    ims[-1] = cv2.cvtColor(ims[-1], cv2.COLOR_BGR2RGB)

#showing images

f, ax = plt.subplots(2,2)
for i in range(4):
    ax[i//2, i%2].imshow(ims[i])
```

```python
# converting image3 to grayscale (using weighted average)

gray_img = 0.299*ims[2][:,:,0] + 0.587*ims[2][:,:,1] + 0.114*ims[2][:,:,2]
plt.imshow(gray_img, 'gray')
plt.show()
cv2.imwrite("../images/gray_img.png", gray_img)
```

```python
# printing dimensions of image2

print("Dimension: ", ims[1].shape)

# Checking the ranges of values in image2

print("Ranges in R channel: Min =", np.min(ims[1][:,:,0]), ", Max =", np.max(ims[1][:,:,0]))
print("Ranges in G channel: Min =", np.min(ims[1][:,:,1]), ", Max =", np.max(ims[1][:,:,1]))
print("Ranges in B channel: Min =", np.min(ims[1][:,:,2]), ", Max =", np.max(ims[1][:,:,2]))
```

This tells us that colour ranges in all 3 channels vary from 0 to 255 and different combinations of R, G, and B produce different colours.
